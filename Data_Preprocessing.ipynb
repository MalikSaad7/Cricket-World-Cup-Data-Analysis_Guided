```python
# importing libraries
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

#convert JSON file into DataFram
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
```

```python
df_match.head()
```

```python
df_match.to_csv('t20_csv_files/Match Summary.csv', index = False)
```

***Creating a Dictionary to link match with match_id***

```python
match_ids_dict = {}

for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

### Batting Summary

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)

all_rec = []
for rec in data:
    all_rec.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_rec)
df_batting.sample(11)
```

***Create a column from dismissal***

```python
df_batting['Out/Not Out'] = df_batting['dismissal'].apply(lambda x: 'Out' if len(x)>0 else 'Not Out')
df_batting
```

***Creating Match_Id Column***

***Removing Unwanted characters***

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)

df_batting.sample(2)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting['dismissal'] = df_batting['dismissal'].apply(lambda x: x.replace('â€', ''))
df_batting['dismissal'] = df_batting['dismissal'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(11)
```

```python
df_batting.to_csv('t20_csv_files/Batting Summary.csv', index = False)
```

### Bowling Summary

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f:
    data = json.load(f)

all_rec = []
for rec in data:
    all_rec.extend(rec['bowlingSummary'])

df_bowling = pd.DataFrame(all_rec)
df_bowling.sample(7)
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)

df_bowling.sample(2)
```

```python
df_bowling['bowlerName'] = df_bowling['bowlerName'].apply(lambda x: x.replace('â€', ''))
df_bowling['bowlerName'] = df_bowling['bowlerName'].apply(lambda x: x.replace('\xa0', ''))
```

```python
df_bowling.to_csv('t20_csv_files/Bowling Summary.csv',index=False)
```

### Players Summary

```python
with open('t20_json_files/t20_wc_player_info.json') as f:
    data = json.load(f)

df_players = pd.DataFrame(data)
df_players.sample(7)
```

```python
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```

```python
df_players.to_csv('t20_csv_files/Players Info no_image.csv', index=False)
```
